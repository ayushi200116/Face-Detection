```python
import cv2 as cv
cascade_face=cv.CascadeClassifier(cv.data.haarcascades +"haarcascade_frontalface_default.xml")
cap=cv.VideoCapture(0)

while True:
    ret,img=cap.read()
    print(ret)
    g=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    f=cascade_face.detectMultiScale(g,1.3,5)
    for(x,y,w,h) in f:
        cv.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
        
    cv.imshow('img',img)
    if cv.waitKey(1) == ord('q'): # press q to stop the show  
        break
cap.release()
cv.destroyAllWindows()
```
